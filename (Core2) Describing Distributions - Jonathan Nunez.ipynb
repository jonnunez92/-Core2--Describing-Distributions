# Load Libraries/Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data/data.csv')
df.head()
```

```python
(df.columns).sort_values()
```

# **VitD_levels**

## 1. Histogram with KDE

```python
col = 'VitD_levels'
feature = df[col]

mean = feature.mean()
median = feature.median()
std = feature.std()

plus1_std = mean + std
minus1_std = mean - std

fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax, stat='probability', kde=True)

ax.axvline(mean, color='red', label=f'Mean = {mean:,.2f}' )
ax.axvline(median, color='green', ls = '--', label=f'Median = {median:,.2f}')
ax.axvline(plus1_std, color = 'black',label=f'+1 std = {plus1_std:,.2f}')
ax.axvline(minus1_std, color = 'black', label = f'-1 std = {minus1_std:,.2f}')
ax.axvspan(plus1_std, minus1_std, color = 'yellow', zorder = 0)

ax.set_title(f'{col}')
ax.legend();
```

```python
from scipy import stats
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt
```

## 2. Questions

1. **Is it Discrete or Continuous?**
    - Continuous since there are no spaces in between bins.
***
2. **Does it have a skew? If so, which direction (+/-)**
    - No Skew, mean and median are pretty much the same
***
3. **What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)**
    - Seems to be Mesokurtic since kurtosis is 3.01
